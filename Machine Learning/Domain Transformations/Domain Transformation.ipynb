# Domain transformation

A domain transformation is described as a mapping $X \to \phi(X)$. This is useful when a different feature space better distinguishes the data distributions. 

Domain transformations can lead to overfitting if the performed transformation is highly specific to the train set but does not generalize well to the real distribution of your classes. Thus, it is advisable to consider the expected distributions of your classes prior to choosing your domain transofrmation. Alternatively, hyperparamter selection techniques can be lent to this purpose. For example cross-validation can help lead to finding a domain transofrmation for a given dataset.

## A domain transformation use case

Let's consider a dataset that is comprised of an inner circle and a ring surrounding it. We can generate this data as follows

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return [x, y]

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return [rho, phi]
```

```python
n = 100
X = np.zeros((n,2))
y = np.zeros((n,))

for i in range(n):
    label = np.random.randint(0, 2)
    if label == 0:
        rho = np.random.uniform(0, 1)
        phi = np.random.uniform(0, 2*np.pi)
        X[i,:] = pol2cart(rho, phi)
    else:
        rho = np.random.uniform(2, 3)
        phi = np.random.uniform(0, 2*np.pi)
        X[i,:] = pol2cart(rho, phi)
    y[i] = label
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='copper')
plt.show()
```

By looking at this data we can see that the two classes are not linearly separable We can see that this

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
reg = LogisticRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
```

```python
# get the separating hyperplane
w = reg.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-3, 3)
yy = a * xx - (reg.intercept_[0]) / w[1]

plt.scatter(X[:,0], X[:,1], c=y, cmap='copper')
plt.plot(xx, yy, color='gray', linewidth=3)
plt.ylim([-4,4])
plt.show()
```

However, when applying a domain transformation, in this case a mapping from the cartesian plane to the polar plane through the equations

$x1 = \sqrt(x1^2 + x2^2)$

$x2 = tan^{-1}(x2/x1)$

We can see that the data in this new feature space is linearly separable.

```python
X_transformed = np.asarray([cart2pol(x, y) for x,y in X])
```

```python
plt.scatter(X_transformed[:,0], X_transformed[:,1], c=y, cmap='copper')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.33)
reg = LogisticRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
```

```python
# get the separating hyperplane
w = reg.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-0.5, 3)
yy = a * xx - (reg.intercept_[0]) / w[1]

plt.scatter(X_transformed[:,0], X_transformed[:,1], c=y, cmap='copper')
plt.plot(xx, yy, color='gray', linewidth=3)
plt.ylim([-4,4])
plt.show()
```
